```python
class node:
    def __init__(self, data):
        self.data = data
        self.next = None
        

class linked_list:
    def __init__(self):
        self.head = None
    
    def append(self, data):
        new_node = node(data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
    
    def print_list(self):
        current = self.head
        while current:
            print(current.data)
            current = current.next
            
    def delete_node(self, key):
        temp = self.head
        if temp is not None:
            if temp.data == key:
                self.head = temp.next
                temp = None
                return
        while temp is not None:
            if temp.data == key:
                break
            prev = temp
            temp = temp.next
        if temp == None:
            return
        prev.next = temp.next
        temp = None
        
    def delete_node_at_position(self, position):
        if self.head is None:
            return
        temp = self.head
        if position == 0:
            self.head = temp.next
            temp = None
            return
        for i in range(position - 1):
            temp = temp.next
            if temp is None:
                break
        if temp is None:
            return
        if temp.next is None:
            return
        next = temp.next.next
        temp.next = None
        temp.next = next
        
    def delete_list(self):
        current = self.head
        while current:
            prev = current.next
            del current.data
            current = prev
        self.head = None
        
        
```

```python
# Create a linked list to represent the grades of a student in a class and their grades

if __name__ == '__main__':
    llist = linked_list()
    
    # Adding grades as (Subject, Grade) pairs
    llist.append(("Math", 85))
    llist.append(("Physics", 90))
    llist.append(("Chemistry", 78))
    llist.append(("English", 92))
    
    print("Student Grades:")
    llist.print_list()

```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.grades = linked_list()  # Each student has their own linked list of grades

    def add_grade(self, subject, grade):
        self.grades.append((subject, grade))

    def print_grades(self):
        print(f"Grades for {self.name}:")
        self.grades.print_list()


if __name__ == '__main__':
    students = linked_list()

    student1 = Student("Alice")
    student1.add_grade("Math", 85)
    student1.add_grade("Physics", 90)

    student2 = Student("Bob")
    student2.add_grade("Math", 78)
    student2.add_grade("Physics", 88)

    student3 = Student("Charlie")
    student3.add_grade("Math", 92)
    student3.add_grade("Physics", 81)

    students.append(student1)
    students.append(student2)
    students.append(student3)

    current = students.head
    while current:
        current.data.print_grades()
        current = current.next
```
