# Activity 1 
 Use of programming language to generate the Recurrence series Write a python program to generate the Fibonacci series with variable input condition.

```python
term=5


def fibseq(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibseq(n-1)+fibseq(n-2)

for i in range(term):
    print(fibseq(i))
```

# Activity 2 
Write a python program to solve a given recurrence relation

```python
import math

def recurrence_relationship(k1, k2, n):
    if n == 0:
        return 1  
    elif n == 1:
        return 1  
    
    return k1 * recurrence_relationship(k1, k2, n - 1) + k2 * recurrence_relationship(k1, k2, n - 2)


k1 = 1
k2 = 1
n = 12  # Term to calculate

F_recursive = recurrence_relationship(k1, k2, n)

print(f"Value from function: F({n}) = {F_recursive}")
```

# Activity 3
Use python language to draw a hash diagram of a given set under divisibility condition.

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

def draw_hasse_diagram(elements):
    elements = sorted(elements)
    
    G = nx.DiGraph()
    G.add_nodes_from(elements)
    
    for i in elements:
        for j in elements:
            if i < j and j % i == 0:
                is_cover = all(k == i or k == j or k % i != 0 or j % k != 0 for k in elements)
                if is_cover:
                    G.add_edge(i, j)
    
    layer_mapping = {node: len([e for e in elements if e % node == 0]) for node in elements}
    for node, layer in layer_mapping.items():
        G.nodes[node]["subset"] = layer  # Add subset attribute for multipartite_layout
    
    pos = nx.multipartite_layout(G, subset_key="subset")
    
    rotation_matrix = np.array([[0, 1], [-1, 0]])  # 90-degree counterclockwise rotation matrix
    for node in pos:
        pos[node] = np.dot(rotation_matrix, pos[node])

    plt.figure(figsize=(10, 8))
    nx.draw_networkx_nodes(G, pos, node_size=700, node_color="lightblue")
    nx.draw_networkx_edges(G, pos, arrows=False)  # No arrows in Hasse diagrams
    nx.draw_networkx_labels(G, pos, font_size=12, font_color="black")
    
    plt.title("Hasse Diagram under Divisibility Condition (Rotated Counterclockwise)")
    plt.axis("off")
    plt.show()

elements = [3, 4, 12, 24, 48, 72]
draw_hasse_diagram(elements)
```
